# Distribuciones de probabilidad

Recordemos lo que es una distribución de probabilidad:

> Una distribución de probabilidad de una variable aleatoria es una función que asigna o asocia probabilidades a los posibles valores que puede tomar la variable.

¿Y qué es una variable aleatorias? Recordemos que:

> Una variable aleatoria es una función que asocia un número real con cada elemento del espacio muestral de un experimento. Dicho de otro modo, una variable aleatoria representa el conjunto de valores que pueden observarse en un experimento, y como los resultados de un experimento están sujetos al azar, sus valores asociados también lo estarán.

Veamos un ejemplo de la ayudantía anterior: La distribución de probabilidades del número de clientes que se atienden en cierta botillería entre las 23:00 y 01:00 horas de un día sábado está dada por la siguiente tabla:

<table>
    <tr><th>Número de clientes</th><th>P(X=x)</th></tr>
    <tr style="text-align: center"><td>3</td><td>0,12</td></tr>
    <tr style="text-align: center;"><td>4</td><td>0,19</td></tr>
    <tr style="text-align: center;"><td>5</td><td>0,28</td></tr>
    <tr style="text-align: center;"><td>6</td><td>0,24</td></tr>
    <tr style="text-align: center;"><td>7</td><td>0,1</td></tr>
    <tr style="text-align: center;"><td>8</td><td>0,07</td></tr>
</table>

Es decir (y de nuevo), en la tabla de arriba, tenemos la distribución de probabilidades de la variable del número de clientes que pueden llegar a cierta botillería. Tenemos que, por ejemplo, la probabilidad de que lleguen $4$ clientes a la botillería, es de $0,19$. O que la probabilidad de que lleguen $6$ clientes o menos es $P(X\le{6}) = P(X=3) + P(X=4) + P(X=5) + P(X=6) = 0,12 + 0,19 + 0,28 + 0,24 = 0,83$.

En la ayudantía pasada estudiamos las distribuciones de probabilidad discretas, las cuales están asociadas a las variables aleatorias discretas. Este tipo de distribuciones cumple con, tal como lo vimos en la semana pasada, las siguientes propiedades:

1. $f(x)\ge0,$
2. $\sum{f(x)}=1$
3. $P(X = x) = f(x)$

De la tercera propiedad desprendemos lo siguiente: en las distribuciones de probabilidad discrteras, **la función de probabilidad nos indica la probabilidad de que la variable aleatoria tome exactamente algún valor en concreto**. Contrario es el caso de las distribuciones de probabilidad continuas, que veremos a continuación.

## Variables aleatorias continuas

Una variable aleatoria continua tiene una probabilidad $0$ de tener exactamente cualquiera de sus valores. Por lo tanto, no podemos representar una variable aleatoria continiua de forma tabular (con una tabla), como en el ejemplo del dado o la botillería. En lenguaje matemático, esto es:

$$P(X=x) = 0 ,\text{para todo x}$$

Recordemos que las variables aleatorias continuas resultan de la medición y pueden tomar infinitos valores dentro de un intervalo (por ejemplo, la estatura, la distancia recorrida, la masa de algún objeto, el tiempo, la temperatura, etc. Cuando trabajamos con distribuciones de probabilidad continua, nos interesará la probabilidad de que la variable aleatorio en cuestión tome valores dentro de cierto rango. Todo lo anterior, Walpole (2012) lo explica muy bien por medio de un ejemplo:

> Consideremos una variable aleatoria cuyos valores son las estaturas de todas las personas mayores de 21 años de edad. Entre cualesquiera dos valores, digamos 163.5 y 164.5 centímetros, o incluso entre 163.99 y 164.01 centímetros, hay un número infinito de estaturas, una de las cuales es 164 centímetros. La probabilidad de seleccionar al azar a una persona que tenga exactamente 164 centímetros de estatura en lugar de una del conjunto infinitamente grande de estaturas tan cercanas a 164 centímetros que humanamente no sea posible medir la diferencia es remota, por consiguiente, asignamos una probabilidad 0 a tal evento. Sin embargo, **esto no ocurre si nos referimos a la probabilidad de seleccionar a una persona que mida al menos 163 centímetros pero no más de 165 centímetros de estatura. Aquí nos referimos a un intervalo en vez de a un valor puntual de nuestra variable aleatoria**.

Como no podemos representar de forma tabular a una distribución de probabilidad continua, usaremos una fórmula para hacerlo: una función $f(x)$ de los valores de la variable aleatoria. A esta función se le denomina **función de densidad de probabilidad** o **función de densidad** de X, la cual tiene las siguientes propiedades:

1. $f(x)\ge0, \text{para todo x}\in{R} $
2. $\int_{-\infty}^{\infty}  f(x) \, dx = 1$
3. $P(a\lt X \lt b) = \int_{a}^{b}  f(x) \, dx$

Lo primera propiedad nos dice que la probabilidad será positiva para cada valor valor en los reales. La segunda propiedad indica que el área bajo la curva definida por la función de densidad es igual $1$. Finalmente, la tercera propiedad señala que la probabilidad de que la variable aleatoria continua $X$ tome un valor entre $a$ y $b$ será la integral definida entre $a$ y $b$ de la función de densidad.

Es decir, **para representar (y calcular) la probabilidades de una variable aleatoria continua, trabajaremos con las áreas definidas bajo la curva de la función de densidad de probabilidad**.

Por ejemplo, en la siguiente figura, vemos la curva generada por cierta función de densidad de probabilidad para una variable aleatoria X. El área sombredada bajo la curva y entre los puntos $a$ y $b$, representa la probabilidad de que la variable aleatoria tome un valor entre $a$ y $b$.

![distribución_continua](img/gráfico_va_continua.png)

Dicha probabilidad se obtiene a través de calcular el valor de la integral $P(a\lt X \lt b) = \int_{a}^{b}  f(x) \, dx$

La **función de distribución acumulativa $F(x)$** de una variable aleatoria $X$ con función de densidad $f(x)$ es:

$$F(x)=P(X \le x) = \int_{-\inf}^{x}f(t) dt$$

Para $-\inf \lt x \lt \inf$.

A continuación veremos una de las familias más importantes dentro de las distribuciones continuas de probabilidad.

### Distribución normal

Es la más importante y de mayor uso de las distribuciones continuas. Se conoce como distribución gaussiana o de Gauss. Se define por dos parámetros,  𝜇  y  𝜎 , que coinciden con la media y la desviación típica. Es decir, toda la familia de distribuciones normales se diferencian por medio de estos dos parámetros. Se suele representar con la siguiente notación $N(\mu, \sigma)$. Cumple con las siguientes características:

* Se representa mediante una curva simétrica conocida como campana de Gauss, en donde el punto más alto de una curva normal se encuentra sobre la media, la cual coincide con la mediana y la moda. La media de una distribución normal puede tener cualquier valor.
![](img/curva_normal.png)
* Las colas de la curva normal se extienden al infinito en ambas direcciones en el eje $X$. 
* Si el valor de la media ($\mu$) cambia, afecta a la posición de la curva normal. Por ejemplo, en la siguiente imagen (tomada de ...), podemos ver 3 curvas normales con 3 medias diferentes: $-10$, $0$, $20$: lo que cambia es su posición sobre el eje de las abscisas.
<img src="img/normales_cambiando_mu.png" width="80%"/>
<br>
* La desviación estándar determina qué tan plana y ancha es la curva normal. Desviaciones estándar grandes corresponden a curvas más planas y más anchas, lo cual indica mayor variabilidad en los datos. Desviaciones estándares de valores pequeñps darán como resultado curvas más altas y "esbeltas", es decir, mayor concentración de datos muy cerca de la media. Por ejemplo, en la siguiente imagen, observamos dos distribuciones normales con la misma medias pero con distinta desviación estándar. La curva negra tiene una desviación estándar mayor que la curva azul, por lo que la primera es más achatada y sus colas más anchas, indicando mayor dispersión de los valores.
<br>
<img src="img/normales_cambiando_sigma.png" width="50%"/>
<br>

La distribución normal rige muchos fenómenos. Las distribuciones de la mayoría de las muestras tomadas en el campo de la industria y la naturaleza se aproximan a la distribución normal, teniendo gran cantidad de aplicaciones prácticas.

La **función de densidad de probabilidad de la distribución normal** es la siguiente:

$$f(x)=\frac{1}{\sqrt{2 \pi \sigma}}e^{-\frac{1}{2\sigma^2}(x - \mu)^2}, -\infty \lt x \lt \infty$$

### Distribución normal estándar

El número de distribuciones normales es ilimitado, y cada una posee diferentes medias, desviación estándar o ambas. normales. Sin embargo, un miembro de la familia de distribuciones normales se utiliza para determinar las probabilidades de
todas las distribuciones de probabilidad normal. Es la **distribución de probabilidad normal estándar** y es única,
pues tiene una media igual $0$ y una desviación estándar igual a $1$. Es decir, en la notación que vimos más arriba, la normal estándar queda denotada como:

$$N(0, 1)$$

Es decir, **una variable aleatoria que tiene una distribución normal con una media cero y desviación estándar de uno tiene una distribución normal estándar**. Para designar esta variable aleatoria normal se suele usar la letra $Z$.

Cualquier distribución de probabilidad normal puede convertirse en una distribución de probabilidad normal estándar al restar la media y cada observación, y dividir esta diferencia entre la desviación estándar. A este proceso se le suele llamar **estandarización**:

$$Z = \frac{x-\mu}{\sigma}$$

El valor $Z$ es la distancia a la media, medida en unidades de desviación estándar. Es decir, **un valor $Z$ expresa la
distancia o diferencia entre un valor particular de la variable aleatoria $X$ y la media aritmética en unidades de desviación estándar**.

#### ¿Y por qué querríamos estandarizar (convertir una distribución normal cualquiera a una distribución normal estándar?

Tal como mencionamos más arriba, el número de distribuciones normales es ilimitado, y cada una posee diferentes medias, desviación estándar o ambas. Recuerda que para calcular probabilidades para una distribución de probabilidad continua debemos calcular áreas bajo la curva de la función de densidad. Para ayudar a calcular las probabilidades, nos podemos ayudar de tablas. Por ejemplo, es posible proporcionar tablas de probabilidad para distribuciones discretas, como la binomial y la de Poisson. Sin embargo, es imposible proporcionar tablas para una infinidad de distribuciones normales que pueden existir. Es por esto que es importante la estandarización: **la distribución normal estándar se utiliza para determinar las probabilidades de todas las distribuciones de probabilidad normal**.

*Ejemplo*:

Dada una distribución normal con $\mu=30$ y $\sigma=6$, determina:

* a) El área bajo la curva normal a la derecha de $x=17$.

Lo primero que debemos hacer, es estandarizar la distribución normal que tenemos. Sabes que su medias es igual a $30$, que su desviación estándar es igual a $6$ y que el valor que queremos estandarizar es $17$. Entonces:

$$Z = \frac{x-\mu}{\sigma}$$
<br>
$$Z = \frac{17-30}{6}$$
<br>
$$Z = -2,17$$

Ahora que sabemos cuánto es $Z$, debemos ir a una **tabla de probabilidades de la normal estándar**, para determinar **cuánto es el área bajo la curva que está "encerrada" desde el $-\inf$ hasta $Z$**, es decir, el área bajo la curva normal estándar desde $-\inf$ hasta $-2,17$.

Lo primero que haremos, es buscar la parte entera y primer decimal de $Z$ en la primera columna de la tabla. En seguida, buscaremos los decimales restantes (desde el segundo), en la primera fila de la tabla:

<br>
<img src="img/ejemplo_tabla_z_1.png" width="70%"/>
<br>

Ahora, intersectamos la fila y columna correspondientes (que encontramos en el paso anterior) para hayar el área bajo la curva que buscamos:

<br>
<img src="img/ejemplo_tabla_z_2.png" width="70%"/>
<br>

Vemos que el número obtenido es $0.0150$. Es decir, el área bajo la curva desde el $-\inf$ hasta el $Z=-2.17$ es $0.0150$. Sin embargo, y como lo hemos señalado, esta es el área a la izquierda de $x=17$. Necesitamos el área a la derecha de aquel valor. Entonces, necesitamos el complemento:

$$P(X>17)=1-0.0150=0.985$$

Por lo tanto, el área bajo la curva normal a la derecha de $x=17$ es $0.985$.

En **R**, tenemos las siguientes funciones para trabajar con la distribución noraml: 

 * ```dnorm```: Función de densidad de probabilidad.
 * ```pnorm```: Función de distribución acumulada.
 * ```qnorm```: Función cuantil de la distribución normal.
 * ```rnorm```: Generación de números pseudoaleatorios según una distribución normal.
 
Mientras que en **Excel** tienes, entre otras, las siguientes fórmulas:

* ```=DISTR.NORM.ESTAND.N(Z; acumulado)```: Devuelve la función de distribución normal estándar (tiene una media de 0 (cero) y una desviación estándar de uno).
* ```=INV.NORM.ESTAND(probabilidad)```: Devuelve el inverso de la distribución normal estándar acumulativa. La distribución tiene una media de cero y una desviación estándar de uno.

## Ejercicios

1. Una emprendedora ha determinado que el tiempo empleado, en horas, en fabricar un determinado producto en su taller sigue una distribución N(10, 2). Es decir, el tiempo empleado es una variable aleatoria $X \sim N(10, 2)$. Calcula la probabilidad de que ese producto se tarde en fabricar:

   a) Menos de 7 horas

    b) Entre 8 y 13 horas

2. Dada una variable aleatoria $X$ tiene una distribución normal con $\mu=300$ y $\sigma=50$, determina la probabilidad de que $X$ tome un valor mayor que 362.

3. La media de una distribución de probabilidad normal es de 60; la desviación estándar es de 5. Con estos datos, calcula:

   a) ¿Alrededor de qué porcentaje de las observaciones se encuentra entre 55 y 65?

b) ¿Cerca de qué porcentaje de las observaciones se encuentra entre 50 y 70?

c) ¿Alrededor de qué porcentaje de las observaciones se encuentra entre 45 y 75?

Del ejercicio anterior, podemos observar que una distribución normal estándar cumple con las siguientes propiedades:

* Cerca del $68\%$ del área bajo la curva normal se encuentra a una desviación estándar de la media (a una desviación estándar a la izquiera y derecha de la media). Esto se puede escribir como ±1σ. Es decir, casi el $68\%$ de las observaciones o valores de la variable aleatoria con distribución normal, quedan encerrados en el área bajo la curva delimitada a una  desviación estándar de la media, tanto a su izquierda como a su derecha.
* Alrededor del $95\%$ del área bajo la curva normal se encuentra a dos desviaciones estándares de la media. Esto se puede escribir como ±2σ. Es decir, aproximadamente el $95\%$ de las observaciones o valores de la variable aleatoria con distribución normal, quedan encerrados en el área bajo la curva delimitada a dos desviaciones estándares de la media, tanto a su izquierda como a su derecha.
* Cerca del $99\%$ de las observaciones se encuentra a tres desviaciones estándares de la media, lo cual se escribe ±3σ.
